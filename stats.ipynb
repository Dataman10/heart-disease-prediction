```python
import numpy as np
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr , 25)
Q1
```

```python
Q2 = np.percentile(arr , 50)
```

```python
Q3 = np.percentile(arr , 75)
Q3
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR
print(lower_fence)
print(upper_fence)
```

```python
lst = []
for i in arr:
    if i>=lower_fence and i<=upper_fence:
        lst.append(i)
```

```python
print(lst)
```

```python
min_element = min(lst)
max_element = max(lst)
```

```python
print(min_element)
print(Q1)
print(Q2)
print(Q3)
print(max_element)
```

```python
arr2 = np.array(lst)
```

```python
print(arr)
print(arr2)
```
