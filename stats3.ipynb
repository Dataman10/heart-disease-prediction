```python
import numpy as np
from scipy.stats import norm
```

```python
from scipy import stats
```

```python
sample = [172 , 174 , 168 , 171 , 173 , 175 , 170 , 169 , 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample , ddof=1)
n = len(sample)
mean_population = 170
```

```python
t_stats = (sample_mean - mean_population)/(sample_std/np.sqrt(n))
t_stats
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_stats), df = n-1))
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis i.e Accept the Alternate Hyptothesis")
else:
    print("Accept the Null Hypothesis")
```
