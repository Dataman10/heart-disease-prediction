```python
def fact(num):
    Ans = 1
    for i in range(1 , num+1):
        Ans *= i
    return Ans

num = int(input("Enter the Element:"))
Ans = fact(num)
print(f"Factorial:{Ans}")
```

```python
def new_fact(num):
    if num<=1:
        return 1
    return num*new_fact(num-1)

num = int(input("Enter the Element:"))
print(new_fact(num))
```

```python
import numpy as np

lst = []
for i in range(1 , 100):
    flag = False
    for j in range(2,int(np.sqrt(i)+1)):
        if i%j == 0:
            flag = True
            break
    if flag == False:
        lst.append(i)
print(lst)
```

```python
def leap_year(year):
    if(year % 4 == 0 or (year % 400 != 0 and year%100 == 0)):
        print("It is a leap year")
    else:
        print("Not a Leap Year")
year = int(input("Enter the Year:"))
print(leap_year(year))
```

```python
def fib(n):
    a , b = 0,1
    for i in range(n):
        print(a , end = " ")
        a , b = b , a+b

num = int(input("Enter the element:"))
fib(num)
```
