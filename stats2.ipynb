```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [19 , 21 , 23 , 18 , 24 , 26 ,17 , 11 , 15 , 22]
n = len(sample)
print(n)
```

```python
population_mean = 18
population_std = 4
sample_mean = np.mean(sample)
```

```python
print(sample_mean)
```

```python
z_score = sample_mean - population_mean/population_std/np.sqrt(n)
print(z_score)
```

```python
p_value = 2*(1 - norm.cdf(abs(z_score)))
print(p_value)
```

```python
alpha = 0.05
if p_value > alpha:
    print("Accept the Null Hypothesis")
else:
    print("Reject the Null Hypothesis")
```
