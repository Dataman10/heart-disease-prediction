```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import norm
```

```python
sample = [172 , 174 , 168 , 169 , 171 , 173 , 175 , 170 , 169,172 , 175 , 178,183, 180,186 , 178 , 182 , 190 , 191,187 , 192,188 , 195 , 196 , 199 , 177 , 189 , 197,190,191]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
print(n)
```

```python
z_score = (sample_mean - population_mean)/population_std/np.sqrt(n)
```

```python
z_score
```

```python
p_value = 2*(1 - norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("I will Reject the Null Hypothesis")
else:
    print("I will Accept the Null Hypothesis")
```
